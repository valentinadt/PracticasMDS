## FORMATO JSON

JSON (JavaScript Object Notation), es un formato para el intercambio de datos basado en texto, muy usado para serializar datos. 



```python
#el primer paso sería cargar las librerías necesarias, incluyendo importar json.

import pandas as pd
import json
```

```python
#importar el data set de iris
data= open('iris.json')
data1 = json.load(data)
data1
```

## FORMATO YAML

El formato YAML (Yet Another Markup Language) sirve para guardar objetos de datos con estructura de árbol. Se dice que es más comprensigle o legible que JSON, el formato comentado anteriormente, además de facilitar el mapeo de estructuras de datos más complejas.

Hace falta una función para leer un fichero YAML.


```python
import yaml

def read_yaml_file(filename):
    with open(filename, 'r') as stream:
        try:
            print(yaml.safe_load(stream))
        except yaml.YAMLError as exc:
            print(exc)
```

En este caso, genero un archivo yaml, para luego poder importarlo.

```python
MostFamousActors_2020 = [{'name': 'Samuel L. Jackson', 'star Score': '350'},
         {'name': 'Dwayne Johnson', 'star Score': '333'},
          {'name': 'Chris Pratt', 'star Score': '299'},
          {'name': 'Tom Holland', 'star Score': '297'}
         ]
```

```python
with open('MostFamousActors_2020.yaml', 'w') as f: #para poder leerlo posteormente
    
    data = yaml.dump(MostFamousActors_2020, f)
```

El siguiente paso sería leer el fichero, con la función predeterminada anteriormente.

```python
read_yaml_file("MostFamousActors_2020.yaml")
```

```python
Para una mejor visualización se pasa a dataframe.
```

```python

actors = pd.DataFrame(MostFamousActors_2020)
print(actors)
```
